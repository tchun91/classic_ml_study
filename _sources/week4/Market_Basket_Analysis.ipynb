## Market Basket Analysis (MBA)

A model that find combinations of items that occur together frequently in transactions and define relationships bewteen the items that people would buy.

## Concepts

**Assoication Rules** <br>
A model to analyze retail basket or transaction data and are intended to identify strong rules discovered in transaction data using measures of interestingness, based on the concept of strong rules.

$$ Rule = If \space an \space item \space A \space occurs, \space then \space item \space B \space also \space occurs \space with \space a \space certain \space probability.  $$
$$ X => Y, where \space X < I, Y < I \space and \space X \cup Y = 0 $$
<br>
<br>
**Apriori Algorithm** <br>
A classical algorithm in data mining, used for mining frequent itemsets and relevant association rules. It is devised to operate on a database containing a lot of transactions, for instance, items brought by customers in a store.

<Dataset>

**Item set**: A collection of one or more items.
$$ I = {i_1, i_2, ..., i_k} $$

**Transaction**: A record of a customer's purchase, containing one or more items.
$$ T = {t_1, t_2, ..., t_3} $$

```python
import pandas as pd

data_list = []
data = pd.DataFrame()

f = open("/content/groceries.csv")

for x in f:
  row = x.replace("\n","").split(",")
  data_list.append(row)

f.close()

for row in range(len(data_list)):
  for col in data_list[row]:
    data.loc[row,col] = 1

data.fillna(0, inplace=True)
```

## Metrics

```python
from itertools import combinations

all_items = data.columns.tolist()
item_pairs = list(combinations(all_items, 2))
```

```python
item_pairs[0:10]
```

**Support** <br>
The number of transactions with both $X$ and $Y$ divided by the total number of transactions.
$$ supp(X => Y) = \frac{|X \cup Y|}{n} $$

```python
def support(X_union_Y_count, total_transaction):
  return X_union_Y_count / total_transactions
```

**Confidence** <br>
The percentage in which Y is bought with X.
$$ conf(x=>y) = \frac{supp(X \cup Y)}{supp(X)} $$

```python
def confidence(X_count, X_union_Y_count):
  return X_union_Y_count / X_count
```

**Lift** <br>
The ratio of the observed support to that expected if $X$ and $Y$ were independent.
$$ lift(X=>Y) = \frac{supp(X \cup Y)}{supp(X)supp(Y)}$$

```python
def lift(X_count, Y_count, X_union_Y_count, total_transaction):
  support_X = X_count / total_transactions
  support_Y = Y_count / total_transactions
  support_union = X_union_Y_count / total_transactions
  return support_union/(support_X * support_Y)
```

**Conviction** <br>
The ratio of the expected frequency that X occurs without Y X and Y were independent divided by the observed frequency of incorrect predictions.
$$ conv(X => Y) = \frac{1-supp(Y)}{1-conf(X=>Y)} $$

```python
def conviction(Y_count, total_transactions, confidence_value):
  support_Y = Y_count / total_transactions
  return (1-support_Y) / (1-confidence_value)
```

**Calculate Metrics**

1. $X$ = Number of transactions containing item X.
2. $Y$ = Number of transactions containing item Y.
3. $X \cup Y$ = Number of transactions containing both X and Y.
4. Total transactions = Total number of rows in the dataset.

```python
def calculate_metrics(data, X, Y):
    X_count = len(data[data[X].all(axis=1)])
    Y_count = len(data[data[Y].all(axis=1)])
    X_union_Y_count = len(data[(data[X] == 1).all(axis=1) & (data[Y] == 1).all(axis=1)])
    total_transactions = len(data)
    return X_count, Y_count, X_union_Y_count, total_transactions
```

Set the thresholds for some metrics (support and confidence).

```python
min_support = 0.01
min_confidence = 0.1
```

- `antecedent`: The item or set of items that appears before the 'then' in an association rule.
- `consequent`: The item or set of items that appears after the 'then' in an association rule.
- `support`: How popular an itemset is, as measured by the proportion of transactions in which an itemset appears. (How frequently a particular combination of items appears together in transactions.)
- `confidence`: The likelihood of item Y being purchased when item X is purchased, expressing the conditional probability of the consequent given the antecedent.
- `lift`: How much more likely item Y is purchased when item X is purchased, while controlling for how popular item Y is in general.
- `Conviction`: The probability that X appears without Y if they were independent with the actual frequency of incorrect predictions.

```python
results_df = pd.DataFrame(columns=["antecedent", "consequent", "support", "confidence", "lift", "conviction"])
```

```python
for X, Y in item_pairs:
    X_count, Y_count, X_union_Y_count, total_transactions = calculate_metrics(data, [X], [Y])

    support_value = support(X_union_Y_count, total_transactions)

    if support_value < min_support:
        continue

    confidence_value = confidence(X_count, X_union_Y_count)

    if confidence_value < min_confidence:
        continue

    lift_value = lift(X_count, Y_count, X_union_Y_count, total_transactions)
    conviction_value = conviction(Y_count, total_transactions, confidence_value)

    results_df = pd.concat([results_df, pd.DataFrame({
        "antecedent": [X],
        "consequent": [Y],
        "support": [support_value],
        "confidence": [confidence_value],
        "lift": [lift_value],
        "conviction": [conviction_value]
    })], ignore_index=True)
```

```python
results_df.sort_values(by='support', ascending=False)
```

A support value of **0.07** for `{whole milk, other vegetables}` means that 7% of all transactions contain both whole milk and other vegetables.

```python
results_df.sort_values(by='confidence', ascending=False)
```

A confidence value of **0.413** for the rule `{margarine} -> {bread}` implies that 41.3% of the times a customer buys milk, they also buy bread.

```python
results_df.sort_values(by='lift', ascending=False)
```

A lift value of **3.04** for the rule `{beef} -> {root vegetables}` indicates that customers who buy milk are 3.04 times more likely to buy bread compared to the average customer.

```python
results_df.sort_values(by='conviction', ascending=False)
```

A conviction value of **1.33** for the rule `{beef} -> {root vegetables}` implies that the rule would be incorrect 1.3 times as often if the association between beef and root vegetables were purely random.
