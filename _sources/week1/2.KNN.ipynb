## K-Nearest Neighbor(KNN) Algorithm
### 1. What is KNN?
   - We are given prior data and it classifies coordinates into groups by features.
   - For a new data point, knn assigns to the nearest neigbor groups.
   - Supervised learning ML algorithm.
   - It does not make any underlying assumptions.
   - Non parametric and easy to implement.
   - High computing power
   - Curse of dimensionality
   - Do not scale
   - No optimization based algorithm unlike linear regression or neural network. (No gradient descent)
### 2. Distance Metrics
   - Distance metric is used to find the nearest group.
   - Euclidean Distance : $ distance(x,X_i) = \sqrt{\sum_{j=1}^{d} (x_j-X_{ij})^2} $
   - Manhattan Distance : $ d(x,y) = \sum_{i=1}^{d} |x_i - y_i| $
   - Minkowski Distance : $ d(x,y) = (\sum_{i=1}^{n} (x_i-y_i)^p)^{\frac{1}{p}} $
### 3. Algorith in detail
   - DistanceToNN=sort(distance from 1st example, distance from kth example)
   - 1. select number of neighbors - K.
     2. Finding Nearest Neighbors using distance function.
     3. Voting for classification or Taking Average for Regression.
        - The class with the most occurences among the neighbors becomes the predicted output. We set the number of neighbors, which is not same as number of classes. N number of neighbor means N-th smallest number and the majority of them will be the final predicted output.
        - The calculated average value becomes the predicted output. Each neighbor picks closest target value in the training data set and it takes the average to get the final value.
        - Large k leads overfitting and small k leads underfitting.

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
iris = fetch_ucirepo(id=53)['data']['original']
  
# data (as pandas dataframes) 
X = iris[[i for i in iris.columns if i not in ['class']]]
y = iris['class']
```

```python
import numpy as np
from scipy import stats
def knn_classifier(input_data,k,lookup_matrix):
    dist_result = np.linalg.norm(np.array(input_data)-np.array(lookup_matrix),axis=1)
    result = y[dist_result.argsort()][:k].mode().values[0]
    return result


```

```python
result_knn = []
for i in np.array(X):
    result_knn.append(knn_classifier(i,3,X))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X,y)
```

```python
sklearn_knn_result = knn.predict(X)
```

```python
import pandas as pd
df = pd.DataFrame({'pred':result_knn,'sklearn_knn':sklearn_knn_result})
df.loc[df.pred != df.sklearn_knn,:]
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(classification_report(result_knn,y))
```
