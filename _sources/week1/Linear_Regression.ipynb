## **Linear Regression**


A supervised machine learning algorithm that establishes the linear relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data.


### Overview


1. **Goal**
- `Primary Objective`: Find the best fit line that can predict the dependent variable $Y$ based on the independent variables $X$.
- `Key Outcome`: Predict $Y$ from an unknown $X$.
2. **Why is Linear Regression Important?**
- `Clarity`: Provides clear coefficients that indicate the impact of each independent variable.
- `Understanding`: Facilitates deeper insight into data dynamics.
- `Simplicity`: Easy to implement and interpret.
- `Foundation`: Forms the basis for more complex algorithms.
- `Assumption Testing`: Validates key assumptions in data analysis.

### Types


1. Based on `Independent` Variables:
  - **A. Simple Linear Regression**: One independent variable.
  - **B. Multiple Linear Regression**: More than one independent variables.
2. Based on `Dependent` Variables:
  - **A. Univariate Linear Regression**: One dependent variable.
  - **B. Multivariate Regression**: Multiple dependent variables.

**1-A. Simple Linear Regression** <br>
$$ y = \beta_0 + \beta_1X$$
- $Y$: Dependent variable
- $X$: Independent variable
- $\beta_0$: Intercept
- $\beta_1$: Slope


Assumptions
1. `Linearity`: $X$ and $Y$ have a linear relationship.
2. `Independence`: Observations are independent.
3. `Homoscedasticity`: Constant variance of errors across levels of $X$.
4. `Normality`: Residuals (the errors between the predicted and actual values) are nomally distributed.

**1-B. Multiple Linear Regression** <br>
$$ y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + ... + \beta_n X_n$$
- $Y$: Dependent variable
- $X_1, X_2, ..., X_n$: Independent variables
- $\beta_0$: Intercept
- $\beta_1, \beta_2, ..., \beta_n$: Slopes


Assumptions
1. `No Multicollinearity`: Independent variables are not highly correlated.
2. `Additivity`: Effects of predictors are additive.
3. `Feature Selection`: Careful selection to avoid overfitting.
4. `Overfitting`: Model complexity should match data complexity.
5. `Multicollinearity`
  - `Correlation Matrix`: High correlations signal multicollinearity.
  - `Variance Inflation Factor (VIF)`: High VIF (>10) suggests multicollinearity.

### Mathematical Concepts


**1. Best Fit Line** <br>
It is objective to minimize the error between predicted and actual values to find the best linear regression equation.

image.png


- **$y_i$: Observed value**<br>
  The actual value observed in the data.

- **$Y_{p_i}$ Predicted value**<br>
  The value predicted by the regression model.

- **$\epsilon_i$: Random error**<br>The difference between the observed value and the predicted value.

- **$\theta_1$: Intercept**<br> The value of $Y$ when $X = 0$; where the regression line crosses the Y-axis.

- **$\theta_2$: Slope**<br> The rate of change of $Y$ with respect to $X$; how steep the regression line is.

- **$x_i$: Input value**<br>  The independent variable or the input value in the regression model.

- **$Y = \theta_1 + \theta_2 X$: Regression Equation (Blue Line)**<br>  The equation that represents the linear relationship between $X$ and $Y$.


**2. Hypothesis Function** <br>
Assumes a linear relationship between $X$ and $Y: \hat{Y} = \theta_1 + \theta_2X$, and the linear regression model updates these parameters to minimze prediction error to find the best fit line, using a cost function.

**3. Cost Function** <br>
Measures how well the model's predictions match the actual data with gradient descent by determining the optimal values for the intercept $\theta_1$ and the coefficient of the input feature $\theta_2$ providing the best-fit line for the given data points.
- **Mean Squared Error (MSE)**: Average squared difference between predicted and actual values.

  - **Step 1: Calculate the Errors** <br>
    Compute the error (or residual) for each data point (the difference between the actual value and the predicted value)
    $$Error_i = y_i - \hat{y_i}$$

  - **Step 2: Square the Errors** <br>
    Square each of the errors calculated in step 1 to ensure all errors are positive.
    $$Squared Error_i = (y_i - \hat{y_i})^2$$

  - **Step 3: Sum the Squared Errors** <br>
    Sum up all the squared errors for all data points.
    $$\sum_{i=1}^{n}(y_i - \hat{y_i})^2$$

  - **Step 4: Calculate the Mean of the Squared Error** <br>
    Take the average of the sum of squared errors by dividing by the number of data points $n$.
    $$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y_i})^2$$

```python
import numpy as np

y_actual = np.random.normal(2.5, 1, 20)
y_pred = np.random.normal(2.5, 1, 20)

print(f"y_actual: {y_actual}")
print(f"y_pred: {y_pred}")

# [Step 1] Calculate the Errors
errors = []
for i in range(len(y_actual)):
  error = y_actual[i] - y_pred[i]
  errors.append(error)

print(f"errors: {errors}")

# [Step 2] Square the Errors
squared_errors = []
for i in range(len(errors)):
  squared_error = errors[i]**2
  squared_errors.append(squared_error)

print(f"squared_errors: {squared_errors}")

# [Step 3 and 4] Calculate the Mean of the Squared Errors
MSE = sum(squared_errors) / len(squared_errors)
print(f"MSE: {MSE}")
```

**4. Gradient Descent** <br>
Optimize algorithm by iteratively updates $\theta_1$ and $\theta_2$ to minimize the **MSE** by moving in the direction of the steepest descent with the outputs of cost functions.
image.png

1. Start with random $\theta_1$ and $\theta_2$ values.
2. Iteratively update the values, reaching minimum cost.
  - $\theta_1 := \theta_1 - \alpha(\frac{\partial J}{\partial\theta_1}) := \theta_1 - \alpha(\frac{1}{n} \sum_{i=1}^{n} 2(\hat{y}_i - y_i))$
  - $\theta_2 := \theta_2 - \alpha(\frac{\partial J}{\partial\theta_2}) := \theta_2 - \alpha(\frac{1}{n} \sum_{i=1}^{n}2(\hat{y_i} - y_i))$
  - Since $\frac{\partial \hat{y_i}}{\partial \theta_1} = 1$, the gradient simplifies, and we use this to update $\theta_1$ to reduce the error.
image.png

- Differentiate the cost function with respect to $\theta_1$ (Same as $\theta_2$).
  - **Step 1: Derivative of the Cost Function**<br>
    Take the partial derivative of the cost function $J$ with respect to $\theta_1$.
    $$\frac{\partial J}{\partial \theta_1} = \frac{\partial}{\partial \theta_1} J(\theta_1, \theta_2) $$
  - **Step 2: Expand the Cost Function**<br>
    The cost function $J$ is typically defined as the mean squared error between the predicted values $\hat{y_i}$ and the actual values $y_i$. So, take the derivative of this function:
    $$\frac{\partial}{\partial \theta_1} \left[ \frac{1}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i)^2 \right]$$
  - **Step 3: Apply the Chain Rule**<br>
    Apply the chain rule to differentiate the squared term.
    $$\frac{1}{n} [\sum_{i=1}^{n} 2(\hat{y}_i - y_i) (\frac{\partial(\hat{y}_i - y_i)}{\partial \theta_1} )]$$
  - **Step 4: Derivative of the Prediction Function**<br>
    Take the derivative of $\hat{y_i} - y_i$ with respect to $\theta_1$ to differentiate the predicted value $\hat{y_i}$, which is given by the hypothesis function.
    $$\frac{1}{n} [\sum_{i=1}^{n} 2(\hat{y}_i - y_i) (\frac{\partial(\theta_1 + \theta_2 x_i - y_i)}{\partial \theta_1} )] $$
  - **Step 5: Simplifying the Expression**<br>
    Since $\theta_1$ is a linear term and does not depend on $x_i$ or $y_i$, the derivative simplifies to 1. The remaining terms cancel out. (Since $\frac{\partial \hat{y_i}}{\partial \theta_1} = 1$.)
    $$\frac{1}{n} [\sum_{i=1}^{n} 2(\hat{y}_i - y_i) (1) $$
  - **Step 6: Simplified Gradient**<br>
    Simplify the expression to obtain the gradient of the cost function with respect to $theta_1$.
    $$\frac{2}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i)$$
  

```python
n = len(y_actual)

# [Step 2] Cost Function J, Mean Squared Error
J = (1/n) * np.sum((y_pred - y_actual) ** 2)

# [Step 3, 4, 5] Gradient Calculation
gradients = (2/n) * (y_pred - y_actual)

# Step 4: Display the computed gradient
print(f"Cost function J: {J}")
print(f"Gradients: {gradients}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(len(gradients)), gradients)
plt.show()
```

References
1. **Chain Rule**: A fundamental concept in calculus used to compute the derivative of a composite function.
$$ \frac{d}{dx}f(g(x)) = f'(g(x))*g'(x)$$
  - Take the derivative of the outer function and then multiplying it by the derivative of the inner function.
2. **Why do we need to differentiate the predicted value?**
  - To understand how changes in $\theta_1$ affect the error or cost function. (Directly informs how to adjust $\theta_1$ to minimize the error.
3. **Why does $(\frac{\partial(\theta_1 + \theta_2 x_i - y_i)}{\partial \theta_1} )$ become 1?**
  - when differentiating with respect to $\theta_1$:
    1. The derivative of $\theta_1$ with respect to $\theta_1$ is 1 (since $\theta_1$ is a linear term.)
    2. The derivative of $\theta_2 x_i$ with respect to $\theta_1$ is 0 because $\theta_2 x_i$ does not contain $\theta_1$.
    3. The derivative of $-y_i$ with respect to $\theta_1$ is also 0 because $y_i$ is independent of $theta_1$.
References
4. $\frac{\partial J}{\partial \theta_1}$: the gradient of the cost function J with respect to the parameter $\theta_1$.
5. $\alpha$: The learning rate that controls how big of a step we take in the direction of the gradient.
  - Too large: might overshoot the minimum of the cost function.
  - Too small: the learning process might be very slow.
6. $\alpha \frac{\partial J}{\partial \theta_1}$: The amount by which we adjust $\theta_1$ in each iteration of gradient descent. (= The step size)

### Metrics


**Mean Squared Error (MSE)** <br>
Measures the average of the squared difference between the acutal values ${y_i}$ and the predicted values $\hat{y_i}$.
- **Formula**
  $$ MSE = \frac{1}{n} \sum_{i=1}^{n}(y_i- \hat{\mathbf{y_i}} )^2$$
  - $n$ = the number of data points.
  - $y_i$ = the actual or observed value for the $i^(th)$ data point.
  - $\hat{y_i}$ = the predicted value for the $i%(th)$ data point.
- **Interpretation**
  - The accuracy of a model's predictions.
  - A lower MSE indicates better model performance.
  - Sensitive to outliers (large errors contribute significantly to the overall score.)

```python
MSE = sum(squared_errors) / len(squared_errors)
print(f"MSE: {MSE}")
```

**Mean Absolute Error (MAE)** <br>
Calculates the average of the absolute difference between the actual and predicted values.
- **Formula**
  $$ MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y_i}|$$
  - $n$ = the number of data points.
  - $y_i$ = the actual values.
  - $\hat{y_i}$ = the predicted values.
- **Interpretation**
  - A lower MAE indicates better model performance.
  - Less sensitive to outliers than MSE, as it does not square the errors.


```python
absolute_errors = []
for i in range(len(errors)):
  absolute_error = abs(errors[i])
  absolute_errors.append(absolute_error)

print(f"absolute_errors: {absolute_errors}")

MAE = sum(absolute_errors) / len(absolute_errors)
print(f"MAE: {MAE}")
```

**Root Mean Squared Error (RMSE)**<br>
Computes the square root of MSE and provides an estimate of how well the observed data points match the predicted values.
- **Formula**
  $$
  RMSE = \sqrt{\frac{RSS}{n}} = \sqrt{\frac{\sum_{i=1}^{n} \left(y_i^{\text{actual}} - y_i^{\text{predict}}\right)^2}{n-2}}
  $$
- **Interpretation**
  - It gives a sense of the typical size of prediction errors in the same units as the dependent variable.
  - The same units as the target variable, making it easier to interpret.
  - Useful for comparing models.
  - Sensitive to the scale of the data.

```python
RMSE = np.sqrt(sum(squared_errors) / (n-2))
print(f"RMSE: {RMSE}")
```

**R-Squared** <br>
Indicates the proportion of the variance in the dependent variable that is explained by the independent variables in the model.
- **Formula**
  $$ R^2 = 1 - (\frac{RSS}{TSS})$$
    - **Residual sum of Squared (RSS)**: The sum of squares of the residual for each data point in the plot or data. (A measurement of the different between the output that was observed and what was anticipated.)
    $$ RSS = \sum_{i=2}^{n} (y_i - b_0 - b_1 x_i)^2$$
    - **Total Sum of Squares (TSS)**: The sum of the data points' errors from the answer variable's mean.
    $$ TSS = \sum(y-\hat(y_i))^2$$
- **Interpretation**
  - Range: from 0 to 1.
  - Closer to 1: a better fit (a high proportion of variance in the dependent variable.)
  - Closer to 0: a worse fit? (a little variance in the dependent variable.)

```python
y_mean = np.mean(y_actual)
tss = sum((y_actual - y_mean)**2)
rss = sum((y_actual - y_pred)**2)

r_2 = 1 - (rss/tss)
print(f"R-squared: {r_2}")
```

**Adjusted R-Squared Error** <br>
Adjusts the r-square value based on the number of predictors in the model, penalizing the inclusion of irrelevant variables.
- **Formula**
  $$ Adjusted R^2 = 1-(\frac{(1-R^2)(n-1)}{n-k-1}$$
  - $n$ = the number of observations
  - $k$ = the number of predictors in the model.
  - $R^2$ = coefficient of determination
- **Interpretation**
  - It accounts for the number of predictors, providing a more accurate measure of model fit, when multiple predictors are used.
  - Higher values indicate a better model after adjusting for the number of predictors.

```python
n = len(y_actual)
p = 1

ar_2 = 1 - ((1 - r_2) * (n - 1) / (n - p - 1))
print(f"Adjusted R-squared: {ar_2}")
```

Reference
1. **When do we use each metric?**
- `MSE`, `RMSE`:  when you need a measure that penalizes larger errors more heavily, and you want to capture the overall fit of the model.
- `MAE`: when you want a straightforward average of errors that is less affected by outliers.
- `$R^2$`: to assess how much of the variance in the dependent variable is explained by the model, especially in simpler models.
- `Adjusted $R^2$`: when comparing models with different numbers of predictors to ensure you're not overfitting by adding too many predictors.
